## üéØ Objetivo
Realizar una **radiograf√≠a estad√≠stica** completa de la Tabla Maestra unificada. En esta fase no buscamos patrones visuales, sino entender la **naturaleza num√©rica** de nuestros clientes: promedios, dispersi√≥n de datos y frecuencias.

## üìÇ Fuente de Datos
* **Archivo:** `../data/procesados/datos_finales_analisis.csv`
* **Contenido:** ~20.000 registros consolidados (Banco + Clientes).

## üîé M√©tricas a Analizar
1.  **Tendencia Central:** ¬øCu√°l es el perfil "promedio" del cliente? (Media y Mediana de edad, ingresos, etc.).
2.  **Dispersi√≥n:** ¬øLos datos son homog√©neos o hay mucha desigualdad? (Desviaci√≥n est√°ndar).
3.  **Frecuencias (Moda):**
    * **Target (`y`):** Tasa base de √©xito de la campa√±a (S√≠ vs. No).
    * **Perfil:** Trabajo, estado civil y educaci√≥n m√°s comunes.

## üõ†Ô∏è Herramientas
* **Pandas:** Para el c√°lculo de estad√≠sticos (`describe`, `mode`, `value_counts`).

```python
# -------------------------------------------------------------------------
# C√ìDIGO: AN√ÅLISIS DESCRIPTIVO
# -------------------------------------------------------------------------
import pandas as pd

# 1. CARGA DE DATOS (Correcci√≥n del error)
# Cargamos el archivo aqu√≠ mismo para asegurar que 'df' existe
df = pd.read_csv('../data/procesados/datos_finales_analisis.csv')

# 2. Resumen Num√©rico (Media, M√≠n, M√°x, Desviaci√≥n)
print("--- üìä ESTAD√çSTICAS VARIABLES NUM√âRICAS ---")
display(df.describe().T)

# 3. An√°lisis de la Variable Objetivo (N√∫meros exactos)
print("\n--- üéØ BALANCE DE LA CAMPA√ëA (Target) ---")
conteo = df['y'].value_counts()
porcentaje = df['y'].value_counts(normalize=True) * 100

# Creamos un peque√±o dataframe para mostrarlo bonito
df_target = pd.DataFrame({'Total': conteo, 'Porcentaje (%)': porcentaje})
display(df_target)

# 4. An√°lisis de Variables Categ√≥ricas (Top valores)
print("\n--- üìã PERFIL T√çPICO (Moda) ---")
cols_cat = df.select_dtypes(include='object').columns

for col in cols_cat:
    # Solo calculamos la moda si la columna existe
    if col in df.columns:
        moda = df[col].mode()[0]
        print(f"-> {col.upper()}: Lo m√°s com√∫n es '{moda}'")
```
